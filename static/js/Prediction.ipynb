```python
from prophet import Prophet
```

```python
sweden_prof_df = sweden_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_male': 'y'})
f_sweden_prof_df = sweden_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_female': 'y'})
sweden_prof_df['ds'] = pd.to_datetime(sweden_prof_df['ds'], format="%Y")
f_sweden_prof_df['ds'] = pd.to_datetime(f_sweden_prof_df['ds'], format="%Y")

sweden_prof_df.head()
```

```python
sweden_model = Prophet()
sweden_model.fit(sweden_prof_df)
f_sweden_model = Prophet()
f_sweden_model.fit(f_sweden_prof_df)
sweden_future = sweden_model.make_future_dataframe(periods=15,freq='YS')
f_sweden_future = f_sweden_model.make_future_dataframe(periods=15,freq='YS')

sweden_forecast = sweden_model.predict(sweden_future)
f_sweden_forecast = f_sweden_model.predict(f_sweden_future)
```

```python
sweden_merged = pd.merge(sweden_prof_df, sweden_forecast[['ds', 'yhat']], on='ds', how='left')
f_sweden_merged = pd.merge(f_sweden_prof_df, f_sweden_forecast[['ds', 'yhat']], on='ds', how='left')


sweden_merged.dropna(inplace=True)
f_sweden_merged.dropna(inplace=True)


r_squared_male = r2_score(sweden_merged['y'], sweden_merged['yhat'])
print("R-squared for Sweden male model:", r_squared_male)


r_squared_female = r2_score(f_sweden_merged['y'], f_sweden_merged['yhat'])
print("R-squared for Sweden female model:", r_squared_female)
```

```python
print(sweden_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20))
sweden_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20).to_csv('sweden_forecast_results.csv', index=False)
files.download('sweden_forecast_results.csv')
```

```python
print(f_sweden_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20))
f_sweden_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20).to_csv('f_sweden_forecast_results.csv', index=False)
files.download('f_sweden_forecast_results.csv')
```

```python
chile_prof_df = chile_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_male': 'y'})
f_chile_prof_df = chile_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_female': 'y'})
chile_prof_df['ds'] = pd.to_datetime(chile_prof_df['ds'], format="%Y")
f_chile_prof_df['ds'] = pd.to_datetime(f_chile_prof_df['ds'], format="%Y")

chile_prof_df.head()
```

```python
chile_model = Prophet()
chile_model.fit(chile_prof_df)
f_chile_model = Prophet()
f_chile_model.fit(f_chile_prof_df)
chile_future = chile_model.make_future_dataframe(periods=15,freq='YS')
f_chile_future = f_chile_model.make_future_dataframe(periods=15,freq='YS')

chile_forecast = chile_model.predict(chile_future)
f_chile_forecast = f_chile_model.predict(f_chile_future)
```

```python
chile_merged = pd.merge(chile_prof_df, chile_forecast, on='ds', how='right')

# Drop any rows with missing values resulting from the merge
chile_merged.dropna(inplace=True)


r_squared_chile_male = r2_score(chile_merged['y'], chile_merged['yhat'])
print("R-squared for Chile male model:", r_squared_chile_male)

f_chile_merged = pd.merge(f_chile_prof_df, f_chile_forecast, on='ds', how='right')


f_chile_merged.dropna(inplace=True)


r_squared_chile_female = r2_score(f_chile_merged['y'], f_chile_merged['yhat'])
print("R-squared for Chile female model:", r_squared_chile_female)
```
