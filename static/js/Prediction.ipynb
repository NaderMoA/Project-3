```python
from prophet import Prophet
```

```python
spain_prof_df = spain_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_male': 'y'})
female_spain_prof_df = spain_df.rename(columns={'Year': 'ds', 'age-standardized_death_rate_per_100k_female': 'y'})
```

```python
spain_prof_df["ds"].dtype
```

```python
spain_prof_df['ds'] = pd.to_datetime(spain_prof_df['ds'], format="%Y")
female_spain_prof_df['ds'] = pd.to_datetime(female_spain_prof_df['ds'], format="%Y")
# max_year = spain_prof_df['ds'].max().year
# start_year = max_year + 1
spain_prof_df.head()
```

```python
spain_prof_df.head()
```

```python
model = Prophet()
model.fit(spain_prof_df)
f_spain_model = Prophet()
f_spain_model.fit(female_spain_prof_df)
```

```python
spain_future = model.make_future_dataframe(periods=15,freq='YS')
f_spain_future = f_spain_model.make_future_dataframe(periods=15,freq='YS')

forecast = model.predict(spain_future)
f_forecast = f_spain_model.predict(f_spain_future)
```

```python
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20))
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20).to_csv('spain_forecast_results.csv', index=False)
files.download('spain_forecast_results.csv')
```

```python
print(f_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20))
f_forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(20).to_csv('f_spain_forecast_results.csv', index=False)
files.download('f_spain_forecast_results.csv')
```
